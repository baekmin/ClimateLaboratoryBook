## 문제 1: 0차원 에너지 균형 모델에서의 시간 의존적 온난화

### 에너지 균형 모델의 방정식

강의에서 우리는 전 지구 평균 지표면 온도 $T_s$에 대한 0차원 에너지 균형 모델을 다음과 같이 정의했습니다.

$$ C  \frac{dT_s}{dt} = \text{ASR} - \text{OLR}$$

$$ \text{ASR} = (1-\alpha) Q $$

$$ \text{OLR} = \tau \sigma T_s^4$$

여기서 우리는 다음 항들을 정의했습니다:

- $C$는 대기-해양 기둥의 열용량입니다.
- $\alpha$는 전 지구 평균 행성 반사율 (albedo)입니다.
- $\sigma = 5.67 \times 10^{-8}$ W m⁻² K⁻⁴는 슈테판-볼츠만 상수입니다.
- $\tau$는 대기에 대한 투과율 매개변수입니다.
- $Q$는 전 지구 평균 입사 태양 복사량, 즉 *일사량 (insolation)*입니다.

매개변수 값은 강의 노트를 참조하십시오.

## 문제 2: 에너지 균형 모델의 알베도 피드백 (Albedo feedback)

### 알베도(Albedo) 공식

이번 실습에서는 **행성 알베도(planetary albedo)가 온도에 의존하도록** 하여 모델에 새로운 물리적 과정을 도입할 것입니다. 그 개념은 행성이 따뜻해질수록 지표면에 얼음과 눈이 줄어들고, 따라서 행성 알베도(반사율)가 낮아진다(반사가 적다)는 것입니다.

빙하-알베도 피드백(ice-albedo feedback)은 다음 공식을 통해 나타낼 수 있습니다:

$$ \alpha(T) = \left\{ \begin{array}{ccc}
\alpha_i &   & T \le T_i \\
\alpha_o + (\alpha_i-\alpha_o) \frac{(T-T_o)^2}{(T_i-T_o)^2} &   & T_i < T < T_o \\
\alpha_o &   & T \ge T_o \end{array} \right\}$$

이때, 다음 매개변수 값들을 사용합니다:

- $\alpha_o = 0.289$는 따뜻하고 얼음이 없는 행성의 알베도입니다.
- $\alpha_i = 0.7$은 매우 차갑고 완전히 얼음으로 덮인 행성의 알베도입니다.
- $T_o = 293$ K는 행성이 얼음이 없다고 가정하는 임계 온도(threshold temperature)입니다.
- $T_i = 260$ K는 행성이 완전히 얼음으로 덮여 있다고 가정하는 임계 온도입니다.

중간 온도에서는 이 공식이 전 지구 평균 온도(global mean temperature)에 따른 알베도의 부드러운 변화를 나타냅니다. 이 공식은 $T = 288$ K일 때 관측된 알베도 $\alpha = 0.299$를 재현하도록 조정되었습니다.
