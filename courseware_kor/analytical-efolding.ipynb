# 심화 주제: 전 지구 에너지 균형 모델의 해석적 해

이 노트북은 올버니 대학교(University at Albany)의 [Brian E. J. Rose](https://www.atmos.albany.edu/facstaff/brose/) 교수가 저술한 [The Climate Laboratory](https://brian-rose.github.io/ClimateLaboratoryBook)의 일부입니다.

## 1. 영차원 에너지 균형 모델: 복습

[이전 강의에서](zero-dim-ebm) 우리는 지배 방정식이 다음과 같은 영차원 **에너지 균형 모델**(Energy Balance Model)을 다루었습니다.

$$ C \frac{dT_s}{dt} = (1-\alpha) Q - \tau \sigma T_s^4 $$

여기서
- $T_s$는 전 지구 평균 표면 온도입니다.
- $C$는 지구 시스템의 **열용량**(heat capacity)으로, 단위는 J m$^{-2}$ K$^{-1}$입니다.
- $\tau$는 대기의 투과율(transmissivity)입니다 (온실 효과의 강도를 나타내는 척도).

우리는 이 시간 의존 모델(time-dependent model)의 수치 해(numerical solutions)가 구현하기 쉽다는 것을 확인했습니다. 하지만 평형 상태로부터의 편차에 대해 OLR(Outgoing Longwave Radiation, 외향 장파 복사)을 선형화하는 (매우 좋은) 근사를 하면 이 모델은 해석적으로 풀 수 있습니다.

해석해(analytical solutions)는 모델에서 실제로 어떤 일이 일어나는지에 대한 상당한 통찰력을 제공할 것입니다.

## 2. 평형 해 주변에서의 선형화

### 평형 해

우리는 이미 이 모델의 평형 해가 다음과 같음을 살펴보았습니다.

$$ T_{eq} = \left( \frac{(1-\alpha) Q}{\tau \sigma} \right)^\frac{1}{4} $$

그리고 이 관계를 바탕으로 모델 매개변수를 조정했습니다.

이 평형 상태에서 벗어나는 작은 섭동(perturbations)에 대해 **방정식을 선형화(linearize)할** 것입니다.

$T_s = T_{eq} + T_s^\prime$라고 가정하고, 우리의 해를 $T_s^\prime << T_{eq}$로 제한합니다.

이는 큰 제약이 아니라는 점에 주목하세요! 예를 들어, 10도(degree)의 온난화(warming) 또는 냉각(cooling)은 절대 평형 온도(absolute equilibrium temperature)의 단지 $\pm$3.4%에 해당합니다.

### 지배 방정식의 선형화

이제 1차 테일러 급수 전개를 이용하여 다음과 같이 나타냅니다.

$$ \text{OLR} = \tau \sigma T_s^4 $$

$$OLR = \tau \sigma T_s^4 = \tau \sigma \left( T_{eq} + T_s^\prime \right)^4  \approx \tau \sigma \left( T_{eq}^4 + 4 T_{eq}^3 T_s^\prime \right) $$

따라서 교란 온도(perturbation temperature)에 대한 수지는 다음과 같이 됩니다.

$$C \frac{d T_s^\prime}{d t} = -\lambda_0 T_s^\prime$$

여기서 우리는 다음을 정의합니다.

$$\lambda_0 = 4 \tau \sigma T_{eq}^3 $$

우리의 관측값을 대입하면, 다음을 얻습니다.

```python
OLRobserved = 238.5  # in W/m2
sigma = 5.67E-8  # S-B constant
Tsobserved = 288.  # global average surface temperature
tau = OLRobserved / sigma / Tsobserved**4  # solve for tuned value of transmissivity

lambda_0 = 4 * sigma * tau * Tsobserved**3
#  This is an example of formatted text output in Python
print( 'lambda_0 = {:.2f} W m-2 K-1'.format(lambda_0)  )
```

이것은 흔히 **플랑크 피드백(Planck feedback)**이라고 불리는 것에 대한 우리의 첫 번째 추정치입니다 (비록 플랑크 피드백 매개변수(Planck feedback parameter)는 더 정확하게는 $-\lambda_0$이며, 이는 음의 피드백(negative feedback)으로 작용하므로 음수입니다).

이는 따뜻한 표면이 우주로 향하는 장파 복사(longwave radiation) 증가를 통해 냉각되려는 경향을 말합니다.

이것은 또한 "피드백 없는(no-feedback)" 기후 반응 매개변수(climate response parameter)라고도 불릴 수 있습니다. 앞으로 보게 되겠지만, $\lambda_0$는 실제 피드백 과정(feedback processes)이 없는 상태에서 기후 시스템(climate system)의 민감도(sensitivity)를 정량화합니다.

## 3. 선형 상미분방정식 풀이

이제 정의합니다.

$$ t^* = \frac{C}{\lambda_0}  $$

이것은 시간(초) 차원을 가지는 양의 상수입니다. 이러한 정의를 바탕으로 온도는 다음과 같이 변화합니다.

$$ \frac{d T_s^\prime}{d t} = - \frac{T_s^\prime}{t^*} $$

이것은 가장 간단한 상미분방정식(ODE) 중 하나입니다. 아마 여러분 대부분에게 친숙할 것입니다. 이것은 **지수 붕괴(exponential decay)** 과정에 대한 방정식입니다.

초기 조건 $T_s^\prime(0)$로부터 적분하여 온도 변화(evolution)를 쉽게 구할 수 있습니다.

$$ \int_{T_s^\prime(0)}^{T_s^\prime(t)} \frac{d T_s^\prime}{T_s^\prime} = -\int_0^t  \frac{dt}{t^*} $$

$$\ln \bigg( \frac{T_s^\prime(t)}{T_s^\prime(0)} \bigg) = -\frac{t}{t^*}$$

$$T_s^\prime(t) = T_s^\prime(0) \exp \bigg(-\frac{t}{t^*} \bigg)$$

저는 이 수업의 모든 학생들이 이 수학적 내용이 직관적이며 이해하기 쉽기를 바랍니다. 그렇지 않다면, 각 단계를 신중하게 검토하여 완전히 이해하시기 바랍니다.

## 4. 전지구 평균 온도의 이완 e-폴딩 시간(e-folding time)

우리 모델에 따르면, 표면 온도는 특성 시간 척도 $t^*$에 걸쳐 평형 값으로 수렴할 것입니다. 이것은 **e-폴딩 시간(e-folding time)**으로, 교란(perturbation)이 $1/e = 0.37$ 배만큼 감쇠(decay)하는 데 걸리는 시간입니다.

*기후 시스템의 경우 이 시간 척도는 얼마여야 할까요?*

$t^*$를 추정하려면 유효 열용량(effective heat capacity) $C$의 값이 필요합니다.

위에서 언급한 우리의 "대략적인(quick and dirty)" 추정치에서는 이 열용량을 결정하기 위해 100미터 깊이의 물을 사용했습니다.

#### 수심 $H$에 대한 적절한 선택은 무엇일까요?

이는 알고 보면 흥미롭고 미묘한 질문입니다. 문제의 시간 척도(timescale)에 크게 좌우됩니다.

- 일 단위?
- 연 단위?
- 수십 년 단위?
- 수천 년 단위?

이 질문은 강좌 후반에 다시 살펴보겠습니다. 우선, $H = 100$ m (해양 표면 혼합층(surface mixed layer)의 일반적인 깊이보다 약간 더 깊은 값)로 가정하고 진행하겠습니다.

이제 표면 온도에 대한 e-폴딩 시간(e-folding time)을 계산해 봅시다:

```python
c_w = 4E3  #  Specific heat of water in J/kg/K
rho_w = 1E3  #  Density of water in kg/m3
H = 100.   #  Depth of water in m
C = c_w * rho_w * H   #  Heat capacity of the model 

tstar = C / lambda_0   #  Calculated value of relaxation time constant
seconds_per_year = 60.*60.*24.*365.
print( 'The e-folding time is {:1.2e} seconds or about {:1.0f} years.'.format(tstar, tstar / seconds_per_year))
```

이것은 행성의 에너지 수지(planetary energy budget)에 영향을 미칠 수 있는 다른 과정들에 비해 상당히 빠른 시간 규모입니다.

**하지만 기후 피드백 매개변수(climate feedback parameter) $\lambda$가 더 작으면, 시간 규모는 더 길어진다는 점에 주목하세요.** 이 내용은 나중에 다시 다룰 것입니다.

## 5. 요약

- 0차원 에너지 균형 모델(EBM)은 외향 장파 복사(OLR)의 온도에 대한 4제곱 의존성을 선형화하여 풀 수 있습니다.
- 그 결과는 간단한 지수 붕괴 모델(exponential decay model)입니다.
- 시스템은 다음 두 가지에 따라 달라지는 $e$-폴딩 시간 척도($e$-folding timescale)로 평형 온도(equilibrium temperature)로 회귀하려는 경향이 있습니다.
    1. 복사 피드백 과정(radiative feedback processes), 그리고
    2. 유효 열용량(effective heat capacity).
- 우리가 추정하기로는, 이 $e$-폴딩 시간은 비교적 짧습니다.
- *열용량을 증가시키거나 피드백 매개변수(feedback parameter)의 (절대)값을 낮출 수 있는 다른 과정들이 없다면, 지구는 에너지 균형(energy balance)에서 크게 벗어나지 않을 것입니다.*
- 학기가 진행됨에 따라 우리는 이 진술을 더 정량화할 것입니다.
