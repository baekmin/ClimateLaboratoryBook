**모델**은

 - 정의하기 쉽지 않습니다!

> 가장 일반적인 의미에서, 모델은 다른 무언가를 어떤 방식으로든 나타내기 위해 사용되는 모든 것을 의미합니다. 어떤 모델은 물리적인 객체입니다. 예를 들어, 조립할 수 있고 심지어 그것이 나타내는 실제 객체처럼 작동하도록 만들어질 수도 있는 장난감 모델이 있습니다. 반면에, 개념 모델(conceptual model)은 개념들의 조합으로 구성된 모델이며, 따라서 오직 정신(마음) 속에만 존재합니다. 개념 모델은 우리가 그것들이 나타내는 대상을 알거나, 이해하거나, 시뮬레이션하는 데 도움을 주기 위해 사용됩니다.

George E. P. Box (통계학자):
> 본질적으로, 모든 모형은 잘못되었지만, 일부는 유용하다.”

From the Climate Modelling Primer, 4th ed {cite:p}`ClimateModellingPrimer:4ed`

> 가장 넓은 의미에서 모델은 세상을 (우리의 경우 기후를) 이해하기 위한 것이며, 그 이해는 모델의 구축(construction)과 조작(manipulation) 과정에서 이루어집니다. 이는 레고로 이상화된 집이나 우주선을 만드는 아이를 보았거나 직접 만들어 본 사람이라면 누구나 알 것입니다. 마찬가지로 기후 모델은 복잡하고 난해한 현실을 이상화하여 표현한 것이며, 이를 통해 기후에 대한 우리의 이해는 크게 확장되었습니다. 모든 모델은 어느 정도 생략, 왜곡, 근사(approximation) 과정을 포함하지만, 점진적으로 모델링되는 시스템에 대한 이해를 구축할 수 있도록 합니다. 아이의 레고 작품은 일반적으로 실제 사물의 필수적인 요소들을 담고 있으며, 세부 사항에 주의를 기울이면 개선되고, 실제 세상을 이해하는 데 도움을 주지만, 결코 실제 사물 그 자체와 혼동되지 않습니다.

## 3. 파이썬을 이용한 우주로의 방출 계산

*이어지는 내용의 대부분은 "빈칸 채우기" 연습으로 기획되었습니다.
*우주로의 장파 복사(longwave emission) 물리적 과정에 대해 논의하면서 파이썬 코드 작성 연습을 할 것입니다.

지구를 유효 전 지구 평균 **방출 온도 $T_e$**를 가진 **흑체 복사체(blackbody radiator)**라고 가정해 봅시다.

그러면 우리는 {term}`외향 장파 복사(Outgoing Longwave Radiation)`를 다음 식으로부터 계산할 수 있습니다.

$$ OLR = \sigma T_e^4 $$

여기서 $\sigma = 5.67 \times 10{-8}$ W m$^{-2}$ K$^{-4}$는 스테판-볼츠만 상수입니다.

**우리는 이것을 방출 온도 $T_e$의 정의로 간주할 수 있습니다.**

관측 자료를 살펴보면, 외향 장파 복사(OLR)의 전 지구 연평균 값은 238.5 W m$^{-2}$입니다.

### 방출 온도 $T_e$ 계산

스테판-볼츠만 법칙을 재정리하면 다음을 얻을 수 있습니다.

$$ T_e = \left(\frac{\text{OLR}}{\sigma} \right)^{\frac{1}{4}} $$

이 모델을 관측값에 맞춰 봅시다:

$$ \tau = \frac{\text{OLR}}{\sigma T_s^4} $$

이것은 매우 편리합니다.
파이썬 콘솔이 당신을 위해 기록을 남겨주고 있습니다!

## 4. 요약

- **기후**는 본질적으로 **날씨의 통계**입니다.
- 지구는 **태양으로부터 흡수되는 에너지**와 **우주로 방출되는 에너지** 간의 **차이**에 반응하여 **따뜻해지거나** **차가워집니다**.
- **기후 모델**은 **지구 시스템**과 **우주** 사이에서 발생하는 이러한 **에너지 교환**을 (수학적 또는 수치적으로) 나타냅니다.
- 관측된 우주 방출량, 즉 {term}`외향 장파 복사(Outgoing Longwave Radiation)`는 255 K의 방출 온도($T_e = 255$ K)와 일치하며, 이는 **지구 표면보다 훨씬 더 낮은 온도**입니다.
- 이것은 **온실 효과**의 증거입니다.
- 우리는 대기에 대한 고정된 *투과율(transmissivity)* $\tau$를 가정하여 매우 단순한 온실 모델을 채택했습니다.
- $\tau$는 개념적으로 ***지표면**에서 방출된 복사 에너지 중 우주로 완전히 도달하는 비율*을 나타냅니다. 이것은 1보다 작은 숫자입니다.
- 파이썬은 재미있고 유용합니다.
