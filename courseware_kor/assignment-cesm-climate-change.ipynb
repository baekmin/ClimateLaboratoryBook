# 과제: CESM 시뮬레이션에서의 기후 변화

이 노트북은 올버니 대학교의 [Brian E. J. Rose](https://www.atmos.albany.edu/facstaff/brose/) 교수가 만든 [The Climate Laboratory](https://brian-rose.github.io/ClimateLaboratoryBook)의 일부입니다.

## 1부

[강의 자료](transient-cesm)의 예시를 따라, 네 가지 CESM 시뮬레이션(완전 결합(fully coupled) 및 슬랩 해양(slab ocean) 버전)을 엽니다.

**전 지구 평균 ASR(흡수된 단파 복사 에너지) 및 OLR(외향 장파 복사 에너지)**의 시계열을 계산하고, 각각을 새로운 변수로 저장하세요. *참고로, CESM 출력에서는 ASR이 `FSNT`로, OLR은 `FLNT`로 불립니다.*

강의 자료에 있는 전 지구 평균 지표면 기온 플롯과 유사하게, **(ASR - OLR), 즉 모델 상단에서의 순 하향 에너지 플럭스**의 시계열을 **12개월 이동 평균(rolling mean)**과 함께 플로팅(plot)하세요.

*참고로, 지표면 기온과 마찬가지로 큰 계절성 주기(seasonal cycle)가 존재하여 우리가 초점을 맞추고자 하는 기후 변화 신호(climate change signal)의 증거를 파악하기 더 어렵게 만들기 때문에 여기서 이동 평균(rolling mean)이 중요합니다.*

## 2부

각 시뮬레이션의 마지막 10년 또는 20년에 대한 **시간 평균 ASR(흡수된 태양 복사 에너지)**과 **시간 평균 OLR(방출 장파 복사 에너지)**을 계산하여 보여주세요. 강의 자료에 따라, 완전 연동 시뮬레이션(fully coupled simulations)에는 20년 구간을 사용하고, 슬랩 오션 시뮬레이션(slab ocean simulations)에는 10년 구간을 사용하세요.

## 3부

1부와 2부의 그림과 수치 결과를 바탕으로 다음 질문에 답하시오:

1.  두 대조군 시뮬레이션(완전 결합(fully coupled) 및 슬랩 오션(slab ocean))은 에너지 균형(energy balance)에 가까운가요?
2.  완전 결합 CO2 점진적 증가 시뮬레이션에서 에너지 불균형(ASR-OLR)은 시간에 따라 증가하나요, 아니면 감소하나요? 80년 시뮬레이션 종료 시점의 불균형 값은 얼마인가요?
3.  슬랩 오션 급격한 2배 CO2 시뮬레이션에 대해 동일한 질문에 답하시오.
4.  완전 결합 시뮬레이션(연 1% CO2 점진적 증가)과 슬랩 오션 시뮬레이션(급격한 2배 CO2)에서 ASR-OLR의 시계열이 매우 다르게 나타나는 이유를 설명하시오. *서로 다른 복사 강제력(radiative forcing)과 서로 다른 해양 열용량(ocean heat capacity)을 모두 고려하여 생각해보세요.*

## 파트 4

CESM 모델에서 이산화탄소(CO2)로 인한 온난화 때문에 전 지구 평균 흡수 단파 복사량(ASR)은 **증가**할까요, 아니면 **감소**할까요?

이를 **양의** 되먹임(positive feedback)으로 설명하시겠습니까, 아니면 **음의** 되먹임(negative feedback)으로 설명하시겠습니까?

## 파트 5

이전 질문에서는 전 지구 평균 흡수 단파 복사(ASR)의 변화를 살펴보았습니다. 이제 세계 여러 지역이 이 변화에 어떻게 기여하는지 살펴보겠습니다.

CO2 강제력(forcing)으로 인한 **흡수 단파 복사(ASR) 변화**에 대한 **지도를 작성**하세요. 결합된 CO2 램프 시뮬레이션의 마지막 20년간의 평균과 제어 시뮬레이션의 마지막 20년간의 평균을 비교하여 사용하세요.

## 6부

5부를 반복하되, 이번에는 ASR(흡수된 태양 복사) 변화 대신 ASR의 **청천(clear-sky)** 성분 변화만 살펴보세요. 이는 `FSNTC`라는 출력 필드에서 찾을 수 있습니다.

*`FSNTC` 필드는 구름이 없을 때의 단파 복사 흡수를 나타내므로, `FSNTC`의 **변화**는 구름 외의 다른 과정에 의해 단파 복사의 흡수 및 반사가 어떻게 영향을 받는지 보여줍니다.*

## 7부

토론:

- 두 지도(흡수 태양 복사량(ASR) 변화, 청천(clear-sky) ASR 변화)가 서로 같게 보이나요?
- 청천(clear-sky) 지도가 왜 그렇게 나타나는지에 대해 몇 가지 아이디어를 제시해 보세요.
- 지도에서 발견한 흥미롭거나, 특이하거나, 놀라운 점에 대해 코멘트해 보세요.
