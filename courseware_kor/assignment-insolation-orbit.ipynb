# 과제: 일사량(Insolation)과 궤도 매개변수(Orbital Parameters)

이 노트북은 Albany 대학교의 [Brian E. J. Rose](https://www.atmos.albany.edu/facstaff/brose/) 저자가 만든 [기후 연구실 (The Climate Laboratory)](https://brian-rose.github.io/ClimateLaboratoryBook)의 일부입니다.

## 지침

강의 노트에 있는 일사량([Insolation](insolation)) 및 궤도 변화([Orbital variations](orbital)) 관련 예시를 참고하여 다음 연습문제를 완료하세요. 마지막 보너스 문제는 필수는 아니지만, 흥미로울 수 있으며 추가 학점을 받을 수 있습니다.

## 1부

다음 세 가지 위도에서 입사 태양 복사 에너지(일사량)를 계산하십시오:
- 적도
- 북위 45º
- 북극

현재의 궤도 매개변수를 사용하십시오.

1. **세 가지 일사량 곡선을 동일한 그림에 모두 표시하며 레이블을 명확히 표시한 그래프를 작성하십시오.** 그래프의 x축은 (1월 1일부터 시작하는) 달력상의 날짜를 나타내야 하며, y축은 W/m2 단위의 일사량이어야 합니다. 각 곡선이 어떤 위도에 해당하는지 나타내는 범례를 포함하십시오.
2. 이 세 곡선의 현저히 다른 형태에 대해 논평하십시오.

## 2부

**10,000년 전의 궤도 매개변수(orbital parameters)를 사용하여 동일한 그래프를 작성하세요** (마지막 빙하기가 끝난 직후). 1부에서 작성한 그래프와 비교하여 다음 질문에 답하십시오:

1. 10,000년 전 북반구 고위도의 하지(summer solstice) 일사량(insolation)은 현재 조건에 비해 약했나요, 아니면 강했나요?
2. 북반구 고위도에서 여름철은 더 길었나요, 아니면 짧았나요? 이를 확인하려면 극지 일출(polar sunrise)과 극지 일몰(polar sunset) 사이의 기간을 살펴보세요.
3. 어떤 다른 차이점들을 발견했나요?

## 3부

남극에서 북극까지 다양한 위도에 대한 **연평균 일사량(annual average insolation)**을 계산하세요.

위도의 함수로서 연평균 일사량을 잘 명시된(well-labeled) 그래프 형태로 결과를 제시하세요. 현재의 궤도 매개변수(orbital parameters)를 사용하거나 다른 어떤 궤도 매개변수를 사용해도 좋지만, 무엇을 사용했는지 명확하게 설명해야 합니다.

## 4부

[우리의 0차원 에너지 균형 모델(EBM)](zero-dim-ebm#equilibrium-temperature)에서 **평형 온도**는 다음과 같음을 상기하십시오.

$$T_{eq} = \left( \frac{(1-\alpha) ~ Q}{\tau~ \sigma} \right)^{\frac{1}{4}}$$

**이 모델이 각 위도에서 독립적으로 적용되며, 여기서 $Q$는 해당 위도에서의 연평균 일사량(Insolation)이라고 가정합시다** (즉, 인접한 위도대 간의 에너지 교환은 무시합니다).

3부에서 계산한 연평균 $Q$를 사용하여, 위도의 함수로 $T_{eq}$의 그래프를 그리십시오. 매개변수 값 $\alpha, \tau$에 대해 당신이 가정한 내용을 명확하게 밝히십시오.

## 파트 5

1. 지구의 남극, 북극, 적도에서 전형적인 **연평균 표면 온도(annual-average surface temperature)** 값은 얼마입니까? 이 수치들에 대한 출처를 반드시 명시하세요.
2. 이 값들은 파트 4에서 계산한 평형 온도와 어떻게 비교됩니까?
3. 파트 4에서 사용한 간단한 모델의 몇 가지 가능한 한계점(shortcomings)에 대해 논의하세요.

## 보너스 질문 (재미 삼아, 추가 점수 획득용)

**이심률(eccentricity)이 0이고 자전축 기울기(obliquity)가 90º인 행성**에 대해 위의 연습문제를 반복하시오.

이 행성의 여러 위치에서 계절별 기온 변화(seasonal cycle of temperature)가 어떻게 나타날지 추측해보시오.

*자전축 기울기(obliquity)가 90º라는 것은 행성의 자전축이 마치 행성이 옆으로 누워 있는 것처럼 지구-태양 평면(earth-sun plane)과 평행하다는 것을 의미합니다. 우리 태양계에서는 천왕성(Uranus)의 자전축 기울기가 90º에 가깝고, 새로 발견된 많은 외계 행성(extra-solar planets)들도 마찬가지입니다.*
