## 서론

[이 강의 노트](elementary-greenhouse)에서 다룬 이층 "새는 온실(leaky greenhouse)" (또는 회색 복사(grey radiation)) 모델을 고려해 봅시다.

여기서는 이 모델을 사용하여 **구름의 복사 효과**를 조사할 것입니다.

구름은 동시에 **단파 복사를 반사**하고 **장파 복사를 흡수**합니다. 이러한 두 가지 효과는 자연에서 종종 서로 상쇄되며, 어떤 효과가 더 강한지는 (다른 요인들과 함께) 구름이 **낮은 구름**인지 **높은 구름**인지(즉, 0층 또는 1층에 있는지)에 따라 달라집니다.

이 문제에서는 (강의 노트에서 그랬듯이) 대기 중에는 **단파 복사의 흡수**가 없다고 가정할 것입니다.

## 질문 1

구름이 위에서 들어오는 단파 복사(shortwave beam) 중 $\alpha_c$만큼을 반사한다고 가정해 봅시다. $\alpha_c$는 0과 1 사이의 값입니다. 구름의 **단파 복사(shortwave)** 효과가 항상 지표면을 **냉각(cooling)**시키는 이유에 대해 일관된 논거를 제시하세요 (글, 그림 및/또는 수식을 사용하여). 구름이 낮거나 높을 때 이 냉각 효과는 달라질까요? 설명하세요.

## 질문 2

구름 속 액체 물방울은 장파 복사(longwave radiation)의 효과적인 흡수체이기 때문에, 구름은 자신이 위치한 층의 **장파 흡수율 / 방출률(longwave absorptivity / emissivity)을 증가시킬 것입니다.**

우리는 2층 대기 모델에서 구름이 있는 층의 흡수율을 $\epsilon + \epsilon_c$로 두어 이를 나타낼 수 있으며, 여기서 $\epsilon_c$는 구름으로 인한 추가적인 흡수율입니다. 온도 $T_s, T_0, T_1$ 및 방출률 $\epsilon, \epsilon_c$를 이용하여 외향 장파 복사(Outgoing Longwave Radiation, OLR)에 대한 공식(즉, 대수적 표현)을 다음 두 가지 경우에 대해 유도하세요:

- 하층운(추가적인 $\epsilon_c$가 0층에 있는 경우)
- 상층운(추가적인 $\epsilon_c$가 1층에 있는 경우)

:::{hint}
각 층에 대해 별도의 방출률 $\epsilon_0, \epsilon_1$을 사용하는 OLR 공식부터 강의 노트에서 시작하세요.

0층의 방출률은 구름이 어디에 있는지에 따라 $\epsilon_0 = \bar{\epsilon}$ 또는 $\epsilon_0 = \bar{\epsilon} + \epsilon_c$가 될 것이며, 여기서 $\bar{\epsilon}$은 우리가 양쪽 층에서 동일하다고 가정하는 정상적인 구름 없는 방출률입니다.
:::

## 문제 3

이제 수업에서 사용했던 조정된 수치들을 사용해 봅시다:

- $T_s = 288$ K
- $T_0 = 275$ K
- $T_1 = 230$ K
- $\epsilon = 0.586$

그리고 $\epsilon_c = 0.2$로 설정합니다.

(a) 다음 질문에 대해 높은 구름과 낮은 구름 각각에 대해 반복하여 답하십시오:

- 구름이 없을 때와 비교하여 구름의 존재로 인한 **OLR (외향 장파 복사) 차이**를 계산하세요.
- 이것이 온난화 효과를 나타내는지, 아니면 냉각 효과를 나타내는지 밝히세요.

(b) 낮은 구름과 높은 구름 중 어떤 것이 더 큰 영향을 미칩니까?

## 질문 4

질문 1-3의 결과를 바탕으로, 하층운(low cloud)과 상층운(high cloud) 중 어느 것이 기후에 **순 온난화 효과(net warming effect)**를 더 유발할 가능성이 높다고 생각하십니까? 말로 설명해 보세요.

## 질문 5
