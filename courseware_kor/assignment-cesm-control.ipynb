## 문제 1: CESM 제어 시뮬레이션의 전 지구 에너지 수지

다음 각 양에 대해 **전 지구 평균, 시간 평균** 값을 계산하고, 이를 강의 자료에 있는 Trenberth and Fasullo (2012) 그림의 관측값과 비교하세요.

- 태양 복사 (단파 복사) 수지:
    - 입사 태양 복사 (Incoming Solar Radiation), 또는 일사량 (Insolation)
    - 대기권 최상단에서 반사된 태양 복사
    - 지표면에 의해 반사된 태양 복사
    - 지표면에 흡수된 태양 복사
    - 구름과 대기에 의해 반사된 태양 복사 *(이 값은 대기권 최상단에서 반사된 복사량과 지표면에서 반사된 복사량의 차이로 계산할 수 있습니다)*
    - 대기권 최상단에서 총 흡수된 태양 복사 (ASR)
    - 대기에 흡수된 태양 복사 *(이 값은 수지의 잔여량, 즉 다른 모든 흡수 및 반사량을 고려한 후 남은 양으로 계산할 수 있습니다)*
- 장파 복사 수지:
    - 외향 장파 복사 (Outgoing Longwave Radiation)
    - 지표면에서 위로 방출되는 복사
    - 지표면으로 향하는 하향 복사
- 기타 지표면 플럭스:
    - "열(Thermals)", 또는 *현열 플럭스 (sensible heat flux)*. *이 값은 데이터셋에서 `SHFLX`라는 필드에서 찾을 수 있습니다.*
    - "증발산(Evapotranspiration)", 또는 *잠열 플럭스 (latent heat flux)*. *이 값은 데이터셋에서 `LHFLX`라는 필드에서 찾을 수 있습니다.*

*참고: 대기 흡수 및 방출 과정에 대해서는 나중에 더 자세히 살펴보겠습니다. "대기 방출(Emitted by Atmosphere)" 또는 "대기창(Atmospheric Window)"과 같은 항을 계산하려고 시도할 필요는 없습니다*

**위 결과를 바탕으로 다음 질문에 답하세요:**

- CESM 제어 시뮬레이션은 **에너지 균형 (energy balance)** 상태에 (또는 그 근처에) 있습니까? 이것이 모델링된 기후가 평형 상태에 가까운지 여부에 대해 무엇을 시사합니까?
- CESM 시뮬레이션의 전 지구 에너지 수지와 관측값 간의 가장 중요한 유사점과 차이점이 무엇이라고 생각하는지 자신의 말로 요약하세요.

## 문제 2: 관측치와 비교하여 전 지구 평균 지표면 온도의 연간 주기 검증

수업 자료에서 우리는 CESM 제어 시뮬레이션의 **전 지구 평균 지표면 온도 시계열**을 플로팅하여 **연간 주기**를 발견했습니다. 이 실습의 목적은 이 현상이 관측된 온도 기록에서도 나타나는지 확인하는 것입니다. 만약 그렇다면, 우리는 이것이 수치 모델의 인공물이 아니라 지구 기후의 실제 특징이라고 결론 내릴 수 있습니다.

관측 자료로는 **NCEP 재분석 데이터(NCEP Reanalysis data)**를 사용할 것입니다.

*재분석 데이터(Reanalysis data)는 실제로 관측 자료와 수치 일기 예보 모델의 출력 결과를 혼합한 것입니다. 이는 관측 자료가 매우 희박한 지역을 포함하여 전 지구의 상태에 대한 "최고의 추정치"를 나타냅니다.*

필요한 데이터는 모두 인터넷을 통해 제공됩니다. 우리는 1981년부터 2010년까지 30년 기간 동안 평균된 월별 기후값(monthly climatologies)을 살펴볼 것입니다.

사용 가능한 데이터는 다음에서 찾아볼 수 있습니다:
https://www.esrl.noaa.gov/psd/data/gridded/data.ncep.reanalysis.derived.html

**지표면 기온(Surface air temperature)**은 `air.2m.mon.ltm.nc`라는 파일에 포함되어 있으며, 이 파일은 `Surface Fluxes`라는 컬렉션에서 찾을 수 있습니다.

이 데이터 파일의 카탈로그 페이지로 바로 가는 링크입니다:
https://www.esrl.noaa.gov/psd/thredds/catalog/Datasets/ncep.reanalysis.derived/surface_gauss/catalog.html?dataset=Datasets/ncep.reanalysis.derived/surface_gauss/air.2m.day.ltm.nc

이제 `OPeNDAP` 링크를 클릭하세요. 파일 내용에 대한 많은 정보가 담긴 페이지가 열릴 것입니다. `Data URL`은 데이터를 파이썬 세션으로 읽어오는 데 필요한 것입니다. 예를 들어, 다음 코드는 파일을 열고 파일에 포함된 변수 목록을 표시합니다:
```python
import xarray as xr
ds = xr.open_dataset('https://www.esrl.noaa.gov/psd/thredds/dodsC/Datasets/ncep.reanalysis.derived/surface_gauss/air.2m.day.ltm.nc')
print(ds)
```

````python
import xarray as xr
url = "http://www.esrl.noaa.gov/psd/thredds/dodsC/Datasets/ncep.reanalysis.derived/surface_gauss/air.2m.mon.ltm.nc"
ncep_air2m = xr.open_dataset(url, decode_times=False)

ncep_air2m
````
