# 기후 시스템과 기후 모델

이 노트북은 올버니 대학교의 [Brian E. J. Rose](https://www.atmos.albany.edu/facstaff/brose/) 교수가 작성한 [The Climate Laboratory (기후 실험실)](https://brian-rose.github.io/ClimateLaboratoryBook)의 일부입니다.

## 1. 기후 시스템 및 그 상호작용

### "기후 시스템"과 그 "권역"의 정의

IPCC AR5 보고서 {cite:p}`AR5_Ch1`에서 발췌:

:::{glossary}
Climate System
: 기후 시스템은 {term}`대기권`, {term}`수권`, {term}`빙권`, {term}`지권`, {term}`생물권`의 다섯 가지 주요 구성 요소와 이들 사이의 상호작용으로 이루어진 고도로 복잡한 시스템이다. 기후 시스템은 자체 내부 역학의 영향뿐만 아니라 화산 폭발, 태양 변동과 같은 외부 강제력(external forcings) 및 대기 조성 변화, 토지 이용 변화와 같은 인위적 강제력(anthropogenic forcings)으로 인해 시간에 따라 진화한다.
:::

이에 따라 몇 가지 추가적인 정의가 필요합니다:

:::{glossary}
Atmosphere
: 지구를 둘러싸고 있는 기체층.

Hydrosphere
: 기후 시스템의 구성 요소 중 액체 상태의 지표수 및 지하수(예: 대양, 바다, 강, 호수, 지하수 등)를 포함하는 부분.

Biosphere
: 모든 생태계와 살아있는 유기체(육상 및 해양 모두)를 포함하는 지구 시스템의 일부… 낙엽, 토양 유기물, 해양 퇴적물과 같이 유래된 죽은 유기물질도 포함합니다.

Cryosphere
: 해빙, 호수 얼음, 강 얼음, 눈 덮개, 빙하 및 빙상, 그리고 얼어붙은 땅(영구 동토층(permafrost) 포함) 등 물이 고체 형태로 존재하는 지구 및 해양 표면 위와 아래의 모든 지역.

Lithosphere
: 대륙과 해양 모두를 포함하는 고체 지구의 최상부 층으로, 모든 지각 암석과 최상부 맨틀의 차갑고 주로 탄성적인 부분을 포함합니다.
:::

### 왜 우리가 이 모든 "권역"들을 모델에 포함시키고 싶어 할까요?

다음은 IPCC AR5 WG1 보고서 {cite:p}`AR5_Ch1`에서 발췌한 두 개의 좋은 그림입니다:

<img src='https://www.ipcc.ch/site/assets/uploads/2017/09/Fig1-01.jpg' width="800">

>그림 1.1 | 기후 변화의 주요 동인. 유입되는 태양 단파 복사(SWR)와 방출되는 장파 복사(OLR) 사이의 복사 평형은 전 지구 기후의 '동인'에 의해 영향을 받습니다. 태양 출력(태양 주기)의 자연적인 변동은 에너지 평형에 변화를 일으킬 수 있습니다 (유입되는 SWR 양의 변동을 통해) (2.3절). 인간 활동은 대기 화학 반응에 관여하는 기체와 에어로졸의 배출량을 변화시켜, 오존(O3)과 에어로졸의 양을 변경합니다 (2.2절). 오존(O3)과 에어로졸 입자는 SWR을 흡수, 산란, 반사하여, 에너지 평형을 변화시킵니다. 일부 에어로졸은 구름 응결핵(cloud condensation nuclei) 역할을 하여 구름 물방울의 특성을 변화시키고 강수량에 영향을 미칠 수 있습니다 (7.4절). 구름이 SWR과 장파 복사(LWR)에 미치는 상호작용이 크기 때문에, 구름 특성의 작은 변화도 복사 수지(radiative budget)에 중요한 영향을 미칩니다 (7.4절). 인간 활동으로 인한 온실가스(GHG, 예: CO2, CH4, N2O, O3, CFCs) 및 대형 에어로졸(크기 >2.5 μm)의 변화는 방출되는 장파 복사(LWR)를 흡수하고 더 낮은 온도에서 더 적은 에너지를 재방출함으로써 방출 LWR의 양을 변화시킵니다 (2.2절). 지표 알베도(albedo)는 식생이나 지표면 특성, 눈이나 얼음 덮개, 해수 색깔의 변화에 의해 변경됩니다 (2.3절). 이러한 변화는 자연적인 계절 및 일주기 변화(예: 눈 덮개)뿐만 아니라, 인간의 영향(예: 식생 유형 변화)에 의해서도 발생합니다 (Forster et al., 2007).

<img src='https://www.ipcc.ch/site/assets/uploads/2017/09/Fig1-02.jpg' width="800">

> 그림 1.2 | 기후 피드백 및 시간 규모. 이산화탄소 증가 및 온도 상승과 관련된 기후 피드백에는 장파 복사(LWR), 기온 감률(lapse rate, 부록 III의 용어집 참조), 그리고 대기-해양 탄소 교환과 같은 음의 피드백(–)과 수증기 및 눈/얼음 반사율(albedo) 피드백과 같은 양의 피드백(+)이 포함됩니다. 일부 피드백은 양(+)이거나 음(–)일 수 있습니다(±): 구름, 해양 순환 변화, 대기-육상 이산화탄소 교환, 그리고 자연계로부터의 비온실가스 및 에어로졸 배출이 그러합니다. 작은 상자 안에는 다양한 피드백의 시간 규모에 대한 큰 차이점이 강조되어 있습니다.

### 핵심은 이 모든 과정들이 궁극적으로 행성 에너지 수지에 영향을 미친다는 것입니다.

**시간 규모(timescales)**에 대해 이야기해 봅시다.

IPCC 자료는 심해 순환과 같이 수백 년의 시간 규모까지만 다룬다는 점에 유의해야 합니다. 하지만 기후 시스템에는 이보다 훨씬 더 긴 시간 규모들이 많이 존재합니다.
예를 들어, 빙상(ice sheets)의 성장과 소멸, 화학적 풍화(chemical weathering)와 같은 지질학적 과정, 대륙 이동(continental drift) 등이 있습니다.

따라서 모델에 어떤 과정을 포함할지에 대한 선택은 **관심 있는 시간 규모**에 따라 결정되어야 합니다. 예를 들어, IPCC 과정은 주로 수백 년의 시간 규모에 초점을 맞추는데, 이는 인간사에 특별한 중요성을 갖기 때문입니다.
그래서 우리는 빙상(ice sheet) 및 지질학적 피드백(feedback)을 모델에 포함하지 않는 경향이 있습니다. 비록 결합 빙상 모델링(coupled ice sheet modeling)이 점점 더 중요해지고 있기는 하지만 말이죠.

## 2. 모의실험 (Simulation) 대 매개변수화 (Parameterization)

### 정의

*기후 모델링 개론* (4판, 124페이지) {cite:p}`ClimateModellingPrimer:4ed`은(는) **매개변수화(parameterization)**를 다음과 같이 정의합니다.

> 물리적 과정을 해당 과정 내의 중요한 매개변수들 사이의 어떤 수학적 관계를 통해 포착하는 과정이지만, 모든 실제 과정을 명시적으로 모델링하지는 않습니다. 예를 들어, 잎의 알베도(albedo, 반사율)는 토양 수분과 관련될 수 있지만 (위조점(wilting point)의 특정 기준을 통해), 잎이 시들 때 식물 내에서 발생하는 물리적 및 생물학적 메커니즘은 모델링되지 않습니다.

우리 연구에서 다룬 한 예시를 들자면, **온실 효과(greenhouse effect)**는 외향 장파 복사(OLR)가 지표면 방출량보다 훨씬 낮게 만들고, 지표면의 평형 온도를 그렇지 않을 때보다 훨씬 더 따뜻하게 만듭니다.

실제로 온실 효과는 다양한 고도에 있는 여러 기체들이 장파 복사(longwave radiation)의 흡수와 방출에 기여함으로써 발생합니다. 주요 온실 기체(greenhouse species)로는 수증기(water vapor)와 구름 방울(cloud droplets)이 있으며, 이들은 모두 날씨의 일일 변화와 밀접하게 관련되어 있습니다.

이를 기후 모델(climate model)에서 어떻게 표현할 수 있을까요?

### 매개변수화

복잡한 복사 전달 과정의 시간 평균된 순 효과와 그것이 일상적인 물 및 구름 변화와 결합되는 방식을 **경험적 관계**를 통해 나타냅니다. 예를 들어,

$$ OLR = \tau \sigma T_s^4 $$

관측된 온도와 OLR을 사용하여 $\tau$에 대한 적절한 값을 선택합니다.

(이것이 우리가 지금까지 해온 것입니다)

#### 장점

구현이 간단하고 결과를 이해하기 쉽습니다.

#### 단점

기후 변화에 따라 $\tau$ 값은 어떻게 변할 수 있을까요? 우리는 알 방법이 없으며, 가정을 해야 합니다.

### 시뮬레이션

대기의 시간 의존적 행동을 설명하기 위해 많은 수의 연립 방정식(coupled equations)을 풉니다:

- 다양한 온실가스에 대한 상세 정보를 포함하는 복사 전달 방정식(Equations of radiative transfer)
- 기상 시스템(weather system)의 발생 및 소멸과 수증기(water vapor)의 이동을 예측하기 위한 유체 운동 방정식(Equations of fluid motion)
- 구름 입자(cloud droplets)의 형성 및 광학적 특성(optical properties)을 예측하기 위한 구름 미세물리(Cloud microphysics)

이것이 소위 GCM(General Circulation Model 또는 Global Climate Model, 전지구 순환 모델 또는 전지구 기후 모델)에서 일어나는 과정의 핵심입니다.

#### 장점

모델은 실제 물리적 원리(예: 뉴턴의 운동 법칙, 질량 및 에너지 보존)에 기반합니다. 따라서 변화하는 기후 조건 하에서도 유효하게 유지될 가능성이 더 높습니다.

#### 단점

많은 컴퓨터 자원이 필요합니다. 우리가 진정으로 원하는 것은 기후(날씨의 통계!)임에도 불구하고 날씨를 모의실험해야 합니다.

본질적으로 **시뮬레이션**은 해당 과정을 지배하는 근본적인 규칙을 (적어도 일부 측면에서) 나타내는 것을 포함합니다. *입력과 출력을 연결하는 인과 관계의 사슬이 존재합니다*.

**매개변수화(Parameterization)**는 해당 과정의 **통계적 특성(statistical properties)**에 대한 가정을 세우는 것을 포함합니다 – 따라서 우리는 실제 사건들을 명시적으로 모델링할 필요 없이, 입력이 주어졌을 때 출력의 몇 가지 관련 통계적 특성을 계산할 수 있습니다.

## 3. 기후 모델의 분류

### 기후 모델의 주요 구성 요소

복사 (Radiation)
: 태양 복사(Solar Radiation)의 유입 및 흡수와 적외선 복사(Infrared Radiation)의 방출을 처리하는 방식

역학 (Dynamics)
: 바람과 해류에 의한 전 지구적 에너지 이동 (특히 저위도에서 고위도로) 및 수직 운동 (예: 소규모 난류, 대류, 심층수 형성 등)

지표면 과정 (Surface processes)
: 해빙 및 육상 빙하, 눈, 식생의 영향과 이로 인한 알베도(Albedo), 방출률(Emissivity)의 변화 및 지표-대기 간 에너지 및 수분 교환을 포함합니다.

화학 (Chemistry)
: 대기의 화학적 조성 및 생물학적 요인을 포함한 다른 구성 요소와의 상호작용 (예: 해양, 육상, 대기 간 탄소 교환)

해상도 (시간 및 공간) (Resolution)
: 모델의 시간 간격(Timestep) 및 모델이 해결하는 수평 및 수직 규모

### 기후 모델링 피라미드

모서리는 모델의 기본 요소를 나타내며, **복잡성은 위로 갈수록 증가합니다**. 피라미드 바닥 주변에는 단 하나의 주요 과정만을 포함하는 더 단순한 기후 모델들이 있습니다.

![입문서 그림 2.2](../images/Primer_Figure2.2.png)

### 기후 모델의 다섯 가지 기본 유형

1. 에너지 균형 모델 (EBMs)
2. 복사-대류 (RC) 또는 단일 칼럼 모델 (SCMs)
3. "차원 제약" 모델
4. 전지구 순환 모델 (GCMs)
5. 지구 시스템 모델 (ESMs)

피라미드 그림의 수직 축은 포함되고 연결되는 과정의 수와 공간적 및 시간적 해상도 측면에서 모두 복잡성이 증가함을 나타냅니다.

### 1. 에너지 균형 모델 (EBM)

다음은 정의입니다:

:::{glossary}
Energy Balance Model
: 지구 에너지 균형(즉, {term}`흡수된 단파 복사`와 {term}`방출 장파 복사` 사이의 불균형)의 함수로 지표면 온도를 예측하는 단순화된 기후 모델(일반적으로 0차원 또는 1차원).

1차원 모델의 경우, 각 위도대에서 에너지 균형에 기여하는 항들을 계산하기 위해 단순화된 관계식이 사용됩니다.

*우리의 0차원 모델을 위도 차원을 포함하도록 어떻게 확장할 수 있을까요? 어떤 새로운 종류의 과정을 포함해야 할까요?*

### 2. 복사-대류 (RC) 또는 단일 기둥 모델 (SCMs)

**수직 기둥** 내의 과정에 중점을 둡니다.

- RC 모델은 **복사 (Radiation)** 및 **대류 (Convection)** 과정을 명시적으로 모델링하여 (일반적으로 전 지구 평균) 연직 온도 프로파일을 계산하며, 이 과정들이 함께 *기온 감률 (Lapse Rate)*을 결정합니다.
- SCM은 3차원 모델에서 '추출된' 단일 기둥으로, 3차원 버전에서 모델링될 모든 과정을 포함하지만, 수평 에너지 전달은 전혀 고려하지 않습니다.

### 3. 차원적으로 제약된 모델

다음 내용을 포함합니다:

-   통계-역학 (Statistical-Dynamical, SD) 모델은 띠 평균 (위도권을 따라 평균한 값) 프레임워크 내에서 표면 과정과 역학을 명시적으로 다루며, 수직적으로 분해된 대기를 가집니다.
-   중간 복잡성 지구 시스템 모델 (Earth-System Models of Intermediate Complexity, EMICs)은 SD 모델을 확장하여 상호작용하는 화학, 특히 해양, 대기 및 육지 사이의 탄소 순환을 포함합니다.
-   통합 평가 모델 (Integrated Assessment Models, IAMs)은 기후 시스템을 경제 활동 모델과 결합하여 배출량에 영향을 미치는 특정 정책 선택의 영향을 더 완전하게 평가합니다.

### 4. 전지구 순환 모델 (GCMs)

:::{glossary}
전지구 순환 모델
: 대기, 해양, 또는 이 둘 모두의 전지구적이고 3차원적인 순환을 모의하는 복잡한 기후 모델.
:::

이러한 모델들은 *완전 연동 해양-대기 모델*의 형태로 존재할 수 있으며, 테스트 및 평가를 위해서는 독립적인 해양 순환 모델 또는 대기 순환 모델의 형태로 존재하기도 합니다.

이 모델들은 가능한 한 많은 과정을 모의하여 전체 기후 시스템의 상태가 시간 경과에 따라 어떻게 변화하는지에 대한 3차원적인 그림을 생성하려고 합니다.

수직 해상도(Vertical resolution)는 수평 해상도(Horizontal resolution)보다 훨씬 더 정밀하지만, 그럼에도 불구하고 층(layers)의 수는 일반적으로 칼럼(columns)의 수보다 훨씬 적습니다.

#### 재미있는 사실

누구에게 묻느냐에 따라, **GCM**은 다음 중 하나를 의미합니다:

- **일반 순환 모델** (General Circulation Model) (원래 의미)
- **전지구 기후 모델** (Global Climate Model) (요즘 더 흔하게 사용됨)
- **전지구 순환 모델** (Global Circulation Model) (아마도 가장 정확한 명칭)

GCM은 원래 대기의 **시간 평균 순환** (time-average circulation)을 형성하고 유지하는 과정을 나타내는 모델이었습니다. 예를 들어, 지표면 바람이 왜 중위도에서는 평균적으로 편서풍이고 열대 지역에서는 편동풍일까요?

역사적으로 **해양** 대순환 모델과 **대기** 대순환 모델 사이에 구분이 있어왔으며, 이들은 각각 **OGCM**과 **AGCM**으로 종종 표기됩니다.

더 복잡한 단계로 나아갈수록, 더 많은 과정들이 통합되어 결합 해양-대기 전지구 모델을 개발하게 되며, 이 모델은 **AOGCM** 또는 **CGCM**으로 표기될 수 있습니다.

### 5. 지구 시스템 모델

피라미드의 더 높은 단계로 나아가고 기후 모델링 역사에서 진전을 이루면서, 과거에는 고정되어 있던 더 많은 과정들이 GCM(전지구 기후 모델)에 통합되고 모델 간의 결합(coupling)이 더욱 완전해집니다. 예를 들어, 생물군계(biomes)와 식생의 변화, 대기, 해양, 토양의 화학적 변화 등이 해당합니다.

이러한 모델들은 이제 일반적으로 **지구 시스템 모델**로 알려져 있습니다.

간결함을 위해, 우리는 기후 시뮬레이션에 사용되는 대기, 해양, 그리고 다른 구성 요소를 포함할 수 있는 모든 복잡한 3차원 모델을 지칭할 때 GCM이라는 일반 용어를 사용합니다. 필요한 경우에는 더 정확하게 설명할 것입니다.

## 4. GCM(전 지구 기후 모델) 소개

### GCM의 구성 요소

GCM의 목표는 대기와 해양의 **완전한 3차원적 특성**을 계산하는 것입니다.

이는 유체 흐름(fluid flow)을 **매개변수화(parameterizing)**하는 대신, 유체 흐름 및 이것이 기후 시스템의 다른 구성 요소에 미치는 영향을 직접 **모의(simulate)**하기 위함입니다.

이를 위해서는 유체(fluid)의 움직임을 지배하는 일련의 방정식들을 풀어야 합니다.

- 에너지 보존 (Conservation of energy)
- 운동량 보존 (Conservation of momentum), 움직이는 유체에 대한 뉴턴의 제2 운동 법칙 ($\vec{F}=m\vec{a}$)
- 질량 보존 (Conservation of mass), 유체 자체(공기 또는 물)는 물론 유체에 의해 운반되는 모든 양(예: 대기 중의 수증기, 해양 중의 염분)에 대한 것
- 상태 방정식 (Equation of state), 유체의 열역학적 특성을 연결하는 것 (예: 대기에 대한 이상 기체 법칙)
- 또한, 구름, 해빙 등의 형성에 대한 방정식

대기과학 전공 학생들은 이 목록, 즉 대기의 '지배 방정식(governing equations)'이라고 불리는 것들을 알아볼 것입니다.

이 방정식들은 *수치 일기 예보(numerical weather prediction)*뿐만 아니라 기후 모델링(climate modeling)의 토대가 됩니다.

물리 법칙은 **미분 방정식**의 형태로 표현됩니다.

우리가 0차원 에너지 균형 모델(EBM)에서 보았듯이, 컴퓨터에서 도함수를 표현하기 위해서는 이들을 근사해야 합니다.

이를 수행하는 한 가지 방법은 지구를 **정규 격자상의 일련의 상자**로 나누는 것입니다.

모델은 대기권에 대한 모든 **상태 변수**를 예를 들어 다음과 같이 나타냅니다.

- 온도
- 풍속 및 풍향
- 습도
- 기압

각 상자에 대해 - 본질적으로 이 양들의 상자 내 공간 평균입니다.

이산화된 방정식은 (적어도 개념적으로는) 우리가 EBM으로 이미 수행했던 것과 같이, **시간 단계별 계산(timestepping)**을 통해 수치적으로 풀립니다.

### 해양 모델의 복잡도 계층

![개론서 그림 2.9](../images/Primer_Figure2.9.png)

이 모든 복잡성의 한 가지 목표는 기후 변화를 더 정확하게 예측하기 위해 **더 많은 시뮬레이션**을 수행하고 **더 적은 매개변수화**를 하는 것입니다.

행성 에너지 수지(planetary energy budget)에 대한 우리의 단순한 관점에서 볼 때, 우리는 **순 기후 피드백 매개변수 $\lambda$를 정확하게 나타내려고 노력하고 있으며**, 이를 통해 올바른 기후 민감도(climate sensitivity)를 얻으려는 것입니다.

이상적으로 이는 모델을 **물리학 및 화학 법칙**에 기반을 두는 것을 의미합니다.

하지만 항상 이런 식으로 작동하는 것은 아닙니다. 많은 경우, 자연에서 피드백(feedback)이 작동한다는 것을 알지만, 그것을 근본 원리(first principles)의 관점에서 표현할 수는 없습니다.

지표면 과정(Land surface processes)이 좋은 예입니다.

육지와 대기 간의 에너지, 물, 탄소 교환은 생물학적으로 매개됩니다. 우리는 (어느 정도) 경험적 관계(empirical relationships)에 의존해야 합니다. 경제 모델링과 다소 비슷합니다.

우리는 또한 공간 규모(spatial scales)에 걸친 상호작용을 다루어야 합니다.

예: 적운 대류(cumulus convection) 및 열과 수증기의 수직 이동(vertical transport).
