# 과제: 복사-대류 모델에서의 되먹임 (Feedbacks)

이 노트북은 올버니 대학교의 [Brian E. J. Rose](https://www.atmos.albany.edu/facstaff/brose/) 저 [The Climate Laboratory](https://brian-rose.github.io/ClimateLaboratoryBook)의 일부입니다.

## 학습 목표

이 과제를 완료하는 학생들은 다음 기술과 개념을 습득하게 될 것입니다:

- climlab을 사용하여 단일 열 복사-대류 모델 (Radiative-Convective Model)을 설정하고 실행하는 방법 익히기
- 기상 스큐-T 차트 (Skew-T charts)에 수직 기온 데이터를 플로팅하고 해석하는 방법 익히기
- climlab을 사용하여 통제된 매개변수 민감도 실험 (parameter-sensitivity experiments) 수행하기
- 기온 감률 되먹임 (lapse rate feedback) 개념 이해
- 복사 강제력 (radiative forcing) 및 기후 되먹임 매개변수 (climate feedback parameters) 계산

## 질문 1: 피드백 없는 기본 복사-대류 모델 (RCM)

여기서는 단일 기둥 복사-대류 모델에서 이산화탄소 (CO$_2$) 농도 두 배 증가의 영향을 살펴봅니다.

*이 연습은 강의 노트에서 다뤘던 내용을 반복하는 것입니다. 과제의 나머지 부분을 수행하기 전에 동일한 결과를 재현할 수 있는지 확인해야 합니다. 아래에서 이 결과들이 필요할 것이기 때문입니다.*

[기후 민감도에 대한 강의 노트](sensitivity-feedback)를 참고하여 다음과 같이 진행하십시오:

- CESM 대조군 (control) 시뮬레이션에서 가져온 비습 (specific humidity)을 사용하여 단일 기둥 복사-대류 모델을 설정하십시오.
- 이 대조군 모델을 평형 상태에 도달할 때까지 실행하십시오.
- 대조군 모델의 복제본 (clone)을 사용하여 대기 중 이산화탄소 (CO$_2$) 농도를 두 배로 늘리고, _대기 조정 복사 강제력_ (atmosphere-adjusted radiative forcing) $\Delta R$을 계산하십시오.
- 이산화탄소 (CO$_2$) 농도가 두 배 증가된 다른 모델 복제본을 사용하여, 모델을 시간 단계별로 평형 상태에 도달할 때까지 실행하십시오. _이 계산에서는 기후가 따뜻해져도 수증기 분포에 변화가 없다고 가정하므로 비습을 수정하지 마십시오._
- 피드백 없는 평형 기후 민감도 (Equilibrium Climate Sensitivity, ECS)를 계산하십시오. _이것은 켈빈 (Kelvin) 단위의 숫자임을 기억하십시오._
- 또한 피드백 없는 기후 반응 매개변수 (climate response parameter) $\lambda_0$를 W m$^{-2}$ K$^{-1}$ 단위로 계산하십시오.

강의 노트에서 얻었던 것과 동일한 결과를 얻었는지 확인하고 보여주십시오.

## 질문 2: RCM에서의 수증기 피드백

*이 질문 역시 강의 노트에서 다루었던 내용을 반복하는 것입니다. 아래의 새로운 질문을 시작하기 전에 동일한 결과를 재현할 수 있는지 확인해야 합니다.*

[기후 민감도에 대한 강의 노트](sensitivity-feedback)를 참고하여 다음을 수행하십시오:

- 기준 RCM에서 상대 습도 프로파일을 계산하십시오.
- 이산화탄소(CO$_2$) 농도를 두 배로 늘린 모델의 새 클론을 만드십시오.
- `for` 루프를 사용하여 모델을 평형 상태까지 시간 진행(timestep)시키십시오. 각 시간 단계(timestep)마다 다음을 수행하여 수증기 피드백을 구현하십시오:
    - 현재 기온을 사용하여 포화 비습도(saturation specific humidity)의 현재 값을 계산하십시오.
    - 상대 습도가 일정하게 유지되도록 비습도(specific humidity)를 수정하십시오.
- 이 모델 버전의 평형 기후 민감도(Equilibrium Climate Sensitivity, ECS)를 계산하십시오.

강의 노트에서 다루었던 것처럼 수증기 피드백을 포함한 이 모델 버전에서 동일한 ECS를 얻는지 확인하고 보여주십시오.

## 문제 3: RCM에서 결합된 기온 감률 및 수증기 피드백

*이제 새로운 작업을 수행할 것이지만, 그 과정은 문제 2에서 수행했던 것과 매우 유사할 것입니다.*

### 지침

지구 온난화의 전형적이고 예상되는 특징은 **대류권 상층부가 지표면보다 더 많이 온난화된다**는 것입니다. (나중에 CESM 시뮬레이션에서 이러한 현상이 실제로 발생함을 확인할 것입니다.)

이러한 특징은 우리의 표준 복사-대류 모델(Radiative-Convective Model, RCM)에는 구현되어 있지 않습니다. 이 모델은 대류 조정(convective adjustment)으로 인해 단일한 고정된 기온 감률(lapse rate)을 따르도록 설정되어 있기 때문입니다.

여기서는 기후가 온난화됨에 따라 다른 물리적 과정들이 이 기온 감률을 변화시킨다고 가정할 것입니다.

**RCM 지구 온난화 계산을 반복하되, 다음 두 가지 피드백을 구현하세요.**

-   **고정된 상대 습도(fixed relative humidity)**를 사용하는 수증기 피드백(즉, 문제 2에서 수행한 내용)
-   다음 공식을 사용하는 **기온 감률 피드백(lapse rate feedback)**:

$$ \Gamma = 6.5 \text{ K} \text{ km}^{-1} - (0.3 \text{ km}^{-1}) \Delta T_s $$

여기서 6.5 K km$^{-1}$는 제어 모델에서 사용했던 임계 기온 감률(critical lapse rate)이며, $\Delta T_s$는 제어 모델 대비 **현재 지표 온난화(surface warming) 값**으로 단위는 K입니다.

예를 들어, 모델이 지표면에서 1 K 온난화되었다면, 우리의 매개변수화(parameterization)에 따르면 임계 기온 감률은 $6.5 - 0.3 \times 1 = 6.2$ K km$^{-1}$가 되어야 합니다.

강의 노트에서 고정된 상대 습도를 구현했던 예시를 따르세요. 각 시간 단계(timestep)마다 복사 과정(radiation process)의 `specific_humidity` 필드를 조정해야 했다는 것을 기억하세요. 여기서도 여전히 이를 수행해야 하지만, 추가적으로 다음 속성(attribute)도 변경해야 합니다.

```
adj_lapse_rate
```

각 시간 단계에서 대류 과정(convection process)의 속성을 변경해야 합니다.

예를 들어, `Convection`이라는 `ConvectiveAdjustment` 과정을 포함하는 `mymodel`이라는 모델이 있다면, 각 시간 단계의 `for` 루프 안에 다음 코드가 필요합니다.
```
mymodel.subprocess['Convection'].adj_lapse_rate = newvalue
```
여기서 `newvalue`는 K/km 단위의 숫자입니다.

문제 2에서와 같이, 각 시간 단계에서 필요한 비습도(specific humidity)를 설정할 뿐만 아니라, 현재 지표 온난화를 사용하여 $\Gamma$의 적절한 현재 값을 계산하고, 위에서 보여준 대로 이 값을 사용하여 `adj_lapse_rate` 매개변수를 설정하는 `for` 루프를 작성해야 합니다.

### 구체적인 질문:

1.  세 가지 온도 연직 분포(temperature profiles)를 보여주는 멋진 skew-T 플롯을 만드세요.
    -   RCM 제어 모델
    -   RCM, 피드백 없는 CO$_2$ 두 배 증가 후 평형 상태
    -   RCM, 수증기 및 기온 감률 피드백이 결합된 CO$_2$ 두 배 증가 후 평형 상태
2.  플롯을 바탕으로, 연직 기둥(column)의 어느 부분에서 가장 큰 온난화가 나타나나요?
3.  수증기 및 기온 감률 피드백이 결합된 새 버전 모델의 평형 기후 민감도(Equilibrium Climate Sensitivity, ECS)를 계산하세요.
4.  이 민감도는 "피드백 없음(no feedback)" ECS보다 크거나 작습니까? 수증기 피드백만 있는 경우의 ECS(강의 노트에서 계산하고 문제 2에서 확인한 값)보다 크거나 작습니까?
5.  (수증기 + 기온 감률) 결합 피드백 매개변수(feedback parameter)를 계산하세요.
6.  이 결과를 강의 노트에 있는 종합 모델(comprehensive models)의 피드백 결과(레이블 "WV+LR")를 담은 IPCC 그림과 비교하세요. 비슷한 숫자를 얻었습니까?
7.  **기온 감률 피드백**을 양의 피드백(positive) 또는 음의 피드백(negative)으로 설명하시겠습니까? 그 이유를 설명하세요.
